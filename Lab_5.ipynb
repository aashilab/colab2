Write a python program to perform **selection sort**.
- ** -Algorithm
* Step 1: Read the number of elements for the list from the user.
* Step 2: Using for loop insert the elements in the list.
* Step 3: Initialize the minimum element as min=numbers[i].
* Step 4: Using the swap method the elements are sorted accordingly.
* Step 5: Print the sorted list.

```python
#write code for above cell and compute the running time and also plot graph.
import time
import random
import matplotlib.pyplot as plt

def selection_sort(numbers):

    n = len(numbers)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if numbers[j] < numbers[min_idx]:
                min_idx = j

        numbers[i], numbers[min_idx] = numbers[min_idx], numbers[i]
    return numbers

def get_user_list():
    while True:
        try:
            num_elements = int(input("Enter the number of elements for the list: "))
            if num_elements <= 0:
                print("Please enter a positive integer.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter an integer.")

    numbers = []

    print(f"Enter {num_elements} elements:")
    for i in range(num_elements):
        while True:
            try:
                element = int(input(f"Element {i+1}: "))
                numbers.append(element)
                break
            except ValueError:
                print("Invalid input. Please enter an integer.")
    return numbers

def measure_running_time(sort_func, data):
    start_time = time.time()
    sort_func(data)
    end_time = time.time()
    return end_time - start_time

def plot_running_time():
    sizes = [10, 50, 100, 200, 500, 1000, 2000, 3000, 4000, 5000]
    times = []
    for size in sizes:
        data = [random.randint(0, 10000) for _ in range(size)]
        time_taken = measure_running_time(selection_sort, list(data))
        times.append(time_taken)
        print(f"Size: {size}, Time: {time_taken:.4f} seconds")

    plt.figure(figsize=(10, 6))
    plt.plot(sizes, times, marker='o', linestyle='-', color='b')
    plt.title('Selection Sort Running Time vs. Input Size')
    plt.xlabel('Input Size (n)')
    plt.ylabel('Time (seconds)')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    user_list = get_user_list()
    sorted_list = selection_sort(list(user_list))
    print(f"\nOriginal list: {user_list}")
    print(f"Sorted list: {sorted_list}")
    time_user_list = measure_running_time(selection_sort, list(user_list))
    print(f"\nRunning time for the user's list ({len(user_list)} elements): {time_user_list:.6f} seconds")
    print("\nGenerating running time graph for various input sizes...")
    plot_running_time()


```

**Quicksort** (sometimes called partition-exchange sort) is an efficient sorting algorithm. Developed by British computer scientist Tony Hoare in 1959 and published in 1961, it is still a commonly used algorithm for sorting.

Quicksort is a divide-and-conquer algorithm. It works by selecting a 'pivot' element from the array and partitioning the other elements into two sub-arrays, according to whether they are less than or greater than the pivot. The sub-arrays are then sorted recursively. This can be done in-place, requiring small additional amounts of memory to perform the sorting.
Write a Python program to sort the elements using quick sort and plot a graph to the time taken versus n.
- Algorithm
1.	Select a pivot element from the array.
2.	Partition the array into two sub-arrays. The elements in the first sub-array are less than the pivot element, while the elements in the second sub-array are greater than the pivot element.
3.	Recursively sort the sub-arrays created in Step 2.
4.	Join the sub-arrays and the pivot element together to	the sorted array.


```python

#write code for above cell and compute the running time and also plot graph.

import time
import random
import matplotlib.pyplot as plt
import sys

sys.setrecursionlimit(2000)

def partition(a, low, high):
    pivot = a[low]
    left = low + 1
    right = high

    while True:
        while left <= right and a[left] <= pivot:
            left = left + 1

        while left <= right and a[right] > pivot:
            right = right - 1

        if left <= right:
            a[left], a[right] = a[right], a[left]
        else:
            break

    a[low], a[right] = a[right], a[low]
    return right

def quicksort(a, low, high):
    if low < high:
        loc_high = partition(a, low, high)
        quicksort(a, low, loc_high - 1)
        quicksort(a, loc_high + 1, high)
def measure_time():
    sizes = [10, 20, 30, 40, 50]
    times = []

    for n in sizes:
        arr = [random.randint(0, 10000) for _ in range(n)]
        start = time.time()
        quicksort(arr, 0, len(arr) - 1)
        end = time.time()
        times.append(end - start)
        print(f"Size: {n}, Time: {end - start:.5f}s")
    plt.figure(figsize=(10, 6))
    plt.plot(sizes, times, marker='o', linestyle='-', color='b')
    plt.title('Quicksort Running Time')
    plt.xlabel('Array Size (n)')
    plt.ylabel('Time (seconds)')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    measure_time()
```
