# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp_logic():
    try:
        n = int(input("Enter the number of cities: "))
    except EOFError:
        return
    print(f"Enter the {n}x{n} matrix (row by row):")
    graph = []
    for i in range(n):
        while True:
            try:
                row_input = input(f"Row {i}: ").split()
                if len(row_input) == n:
                    graph.append([int(x) for x in row_input])
                    break
                else:
                    print(f"Error: Please enter exactly {n} numbers.")
            except ValueError:
                print("Error: Please enter valid integers.")
    dp = [[-1] * (1 << n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1 << n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        res = float('inf')
        for next_city in range(n):
            if not (mask & (1 << next_city)):
                new_cost = graph[city][next_city] + visit(next_city, mask | (1 << next_city))
                res = min(res, new_cost)

        dp[city][mask] = res
        return res
    result = visit(0, 1)
    print("\nMinimum travelling cost:", result)

tsp_logic()

```
